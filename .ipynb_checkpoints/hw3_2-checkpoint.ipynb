## 資料前處理
### 把資料吃吃吃進去！好吃！

```python
import numpy as np 
import pandas as pd
dataset = pd.read_csv('./online_shoppers_intention.csv')
dataset.head(5)
```

### 定義 input（x） 跟 output（y）

```python
X_train = dataset.iloc[:, 0:len(dataset.iloc[[0],:])-2] #x捨棄Survived , .value的話會捨棄掉pclass那些
Y_train = dataset.iloc[:, [-1]] #y是survived
                       
print(X_train.head(5))
print("==================")
print(Y_train.head(5))
```

### 如果特徵中有string，將它轉換成數字

```python
X_train['month'] = 0
for i in range(len(X_train)):
    if X_train['Month'][i] == 'Jan':
        X_train['month'][i] = 1
    elif X_train['Month'][i] == 'Feb':
        X_train['month'][i] = 2
    elif X_train['Month'][i] == 'Mar':
        X_train['month'][i] = 3
    elif X_train['Month'][i] == 'Apr':
        X_train['month'][i] = 4
    elif X_train['Month'][i] == 'May':
        X_train['month'][i] = 5
    elif X_train['Month'][i] == 'Jun':
        X_train['month'][i] = 6
    elif X_train['Month'][i] == 'Jul':
        X_train['month'][i] = 7
    elif X_train['Month'][i] == 'Aug':
        X_train['month'][i] = 8
    elif X_train['Month'][i] == 'Sep':
        X_train['month'][i] = 9
    elif X_train['Month'][i] == 'Oct':
        X_train['month'][i] = 10
    elif X_train['Month'][i] == 'Nov':
        X_train['month'][i] = 11
    elif X_train['Month'][i] == 'Dec':
        X_train['month'][i] = 12

X_train = pd.get_dummies(data=X_train, columns=['Month', 'VisitorType', 'Weekend'])
X_train.head(10) 
```
